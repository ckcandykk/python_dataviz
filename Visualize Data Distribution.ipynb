### scatter plot การ plot จุดที่กระจัดกระจาย

```python
irisdata = pd.read_csv('C:\\Users\\CANDY\\Documents\\python_viz\\python_dataviz\\iris.data' ,header = None) 
irisdata 
```

```python
setosa_np = irisdata.iloc[:50,:-3].to_numpy() #[เลือกแถว,เลือกหลัก]
setosa_np
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy() #[เลือกแถว,เลือกหลัก]
setosa_np
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2]) #plot ตามขนาดของ array หลักที่ 3
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c = 'c') #r g b
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = setosa_np[:,3],cmap='YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = np.array([[1,0.5,0.8]]))
```

plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = setosa_np[:,3],cmap='YlOrBr',marker = "*")

### plot data 2 ตัวในกราฟเดียวกัน
